```python
from transformers import pipeline
```

```python
a = pipeline("sentiment-analysis")
```

```python
text = input("Enter the text: ")
print(a(text))
```

```python
gen = pipeline('text-generation')
```

```python
text = input("Enter the text: ")
t = gen(text, max_length = 200)
```

```python
print(t[0].get('generated_text'))
```

```python
n = pipeline('ner')
```

```python
text = input("Enter a text")
result = n(text)
result
```

```python
qa = pipeline('question-answering')
```

```python
c =  '''Bitty bought some butter, but the butter was bitter butter.
so, Bitty bought some better butter, to make bitter butter better'''
```

```python
q = "what does Bitty bought?"
```

```python
result = qa({'question': q, 'context': c})
print(result)
```

```python
s = pipeline("summarization")
```

```python
t = '''A paragraph is a group of sentences that work together to express a single, unified idea or point.
It is a fundamental building block of writing, used to organize and present information in a clear and coherent manner.
Paragraphs typically start on a new line and are indented, visually signaling a shift in the flow of thought.
Each paragraph should focus on a single main idea or controlling idea.
The sentences within a paragraph should logically connect and flow together, building upon the main idea.'''
```

```python
result = s(t)
result[0].get('summary_text')
```
